```python
import pandas as pd

def text_correct(text):
    text = text.lower()
    return text.replace(' ', '_')
```

```python
b = pd.read_csv('bookings.csv', sep = ';')
```

```python
b = b.rename(columns=text_correct)
```

```python
b.country.value_counts().head(5)
```

```python
succes = b.query('is_canceled == 0')
```

```python
succes.groupby('country', as_index=False) \
.agg({'is_canceled' : 'count'}) \
.sort_values('is_canceled', ascending=False) \
.rename(columns = {'is_canceled' : 'succes_number_reservation'}).head(5)
```
